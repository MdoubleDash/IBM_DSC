Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

## find distance of coastline from CCAFS LC-40
CCAFSLC_40_first_row = spacex_df[spacex_df['Launch Site'] == 'CCAFS LC-40'].iloc[0]
coastline_coord = [28.56307, -80.56798]
distance_coastline = calculate_distance(CCAFSLC_40_first_row.Lat, CCAFSLC_40_first_row.Long, coastline_coord[0], coastline_coord[1])

print("Approximate distance of CCAFS LC-40 site from coastline is", round(distance_coastline,2), "km")
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

coast_marker = folium.Marker(
   coastline_coord,
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#000000;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
       )
   )
site_map.add_child(coast_marker)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

## saving each marker cord to a list
coastline_coord = [28.56307, -80.56798] ## Atlantic Ocean Coastline
town_coord = [28.40106, -80.60532] ## Cape Canaveral (not a major city)
city_coord = [28.10621, -80.64308] ## Melbourne (relatively large city)
HW_coord = [28.56288, -80.57074] ## Samuel C Phillips Parkway
RR_coord = [28.57206, -80.58527] ## NASA Railroad

##calcualte dist for each marker from CCAFS LC-40 
distance_coastline = calculate_distance(CCAFSLC_40_first_row.Lat, CCAFSLC_40_first_row.Long, coastline_coord[0], coastline_coord[1])
distance_town = calculate_distance(CCAFSLC_40_first_row.Lat, CCAFSLC_40_first_row.Long, town_coord[0], town_coord[1])
distance_city = calculate_distance(CCAFSLC_40_first_row.Lat, CCAFSLC_40_first_row.Long, city_coord[0], city_coord[1])
distance_hw = calculate_distance(CCAFSLC_40_first_row.Lat, CCAFSLC_40_first_row.Long, HW_coord[0], HW_coord[1])
distance_rr = calculate_distance(CCAFSLC_40_first_row.Lat, CCAFSLC_40_first_row.Long, RR_coord[0], RR_coord[1])

##create a list for all the coordinates and all the distances to later loop through them
coordinates = [coastline_coord, town_coord, city_coord, HW_coord, RR_coord]
distances = [distance_coastline, distance_town, distance_city, distance_hw, distance_rr]
```

```python
## loop over list of coordinates and add markers
for i, val in enumerate(coordinates):
    proximity_markers = folium.Marker(
        coordinates[i],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#000000;"><b>%s</b></div>' % "{:10.2f} KM".format(distances[i]),
        )
        )
    site_map.add_child(proximity_markers)
    
site_map    
```
