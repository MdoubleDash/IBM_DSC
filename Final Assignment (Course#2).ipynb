### This code aims to create a pandas dataframe and then graph it as a scatter plot

````python
# loading pandas
import pandas as pd

# craeting a list to be converted to a df later
data = {'Age':    [25, 30, 28, 40, 55, 65, 32],
        'Income': [78, 110, 95, 90, 105, 85, 100]}

# converting list to pandas DF
df = pd.DataFrame(data)

# printing dataframe
print(df)

#creating the scatter plot
p = df.plot.scatter(x='Age', y='Income',
                    color='Blue')

#printing an empty line & the plot
print("\n")
print(p)
````
